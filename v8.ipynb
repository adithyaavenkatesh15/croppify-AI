## Test phase

```python
from collections import Counter
```

```python
image_path = r"C:\Users\mohan\Downloads\stamp-doc4.jpg"
image = cv2.imread(image_path)
results = model(image)


confidence_threshold = 0.5
detected_classes = []

for r in results:
    boxes = r.boxes
    labels = r.names  # dict: {class_id: class_name}

    for box in boxes:
        # Skip boxes below confidence threshold
        if box.conf[0] < confidence_threshold:
            continue

        x1, y1, x2, y2 = box.xyxy[0]
        x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)

        class_id = int(box.cls[0])
        class_name = labels.get(class_id, "Unknown")
        detected_classes.append(class_name)

        # Draw bounding box
        cv2.rectangle(image, (x1, y1), (x2, y2), (0, 0, 255), 2)

        # Draw class name above top-right of the box
        text_position = (x2, max(y1 - 10, 0))
        cv2.putText(image, class_name, text_position, cv2.FONT_HERSHEY_SIMPLEX,
                    0.9, (0, 255, 0), 2, cv2.LINE_AA)

# Handle case when nothing is detected
if not detected_classes:
    print("\nNo items detected above confidence threshold.")
else:
    # Print all detected class names
    print("\n[Detected Items]")
    for item in detected_classes:
        print(f"- {item}")

    # Print count of each class
    print("\n[Count of Each Class]")
    class_counts = Counter(detected_classes)
    for class_name, count in class_counts.items():
        print(f"{class_name}: {count}")
```
